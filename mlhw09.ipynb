```python
import numpy as np
import matplotlib.pyplot as plt

p_values = [0.1, 0.5, 0.9]
q_values = np.linspace(0, 1, 100)  # Varying q from 0 to 1

for p in p_values:
    cross_entropy = p * np.log(q_values) + (1 - p) * np.log(1 - q_values)
    plt.plot(q_values, cross_entropy, label=f'p = {p}')

plt.xlabel('q')
plt.ylabel('Cross Entropy')
plt.title('Cross Entropy for Bernoulli Distributions')
plt.legend()
plt.grid(True)
plt.show()
```
