```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def g(z):
    return -np.log(sigmoid(z))

z_values = np.linspace(-5, 0, 30)
g_values = g(z_values)

plt.plot(z_values, g_values)
plt.xlabel('z')
plt.ylabel('g(z)')
plt.title('Plot of g(z) = -log(σ(z))')
plt.grid(True)
plt.show()
```

```python
import numpy as np

Q = np.array([10, 11, 12, 13])  # action values

def softmax(Q, T):
    return np.exp(Q / T) / np.sum(np.exp(Q / T))

# (a) Low temperature, T=1
p_low = softmax(Q, T=1)
print("Low temperature probabilities:", p_low)

# (b) Intermediate temperature, T=10
p_intermediate = softmax(Q, T=10)
print("Intermediate temperature probabilities:", p_intermediate)

# (c) High temperature, T=100
p_high = softmax(Q, T=100)
print("High temperature probabilities:", p_high)
```
